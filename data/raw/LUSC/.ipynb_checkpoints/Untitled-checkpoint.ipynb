```python
import pandas as pd


def genes(df1, df2):
    """
    df2 根据 df1 的基因数据对齐。
    :param df1:
    :param df2:
    :return:
    """
    align_genes = df1.index
    remove_genes = set(df2.index) - set(df1.index)
    df2 = df2.drop(remove_genes)
    return df2


def patients(df1, df2):
    """
    df2 根据 df1 的基因数据对齐。 
    :param df1: 
    :param df2: 
    :return: 
    """
    df1.columns = [patient[0:16] for patient in df1.columns]
    print(df1.columns)
    print(df2.columns)
    remove_patients = set(df2.columns) - set(df1.columns)

    print(remove_patients)
    df2 = df2.drop(remove_patients, axis=1)
    return df2


def select_stage(df, stage=1):
    """
    select specified tumor-stage samples
    :param df:
    :return:
    """
    try:
        remove_patients = [patients for patients in df if df[patients]['cancer'] != stage]
        df = df.drop(remove_patients, axis=1)
        df = df.drop(['cancer'])
    except Exception as e: 
        pass
    return df

```

```python
methy_path = 'methy.txt'
expr_path = 'expr.txt'
    
methy_df = pd.read_csv(methy_path, sep='\t', index_col=0, low_memory=False)
expr_df = pd.read_csv(expr_path, sep='\t', index_col=0, low_memory=False)
expr_df = genes(methy_df, expr_df)  # 按照 methy 的 gene 对齐 expr 的
expr_df = patients(methy_df, expr_df)
methy_df = select_stage(methy_df, 1)
expr_df = select_stage(methy_df, 1)
    
```
